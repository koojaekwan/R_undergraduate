# Summary

## EDA : 탐색적 데이터 분석


- 다양한 기법(그래프)을 활용, 데이터에 대한 인사이트 극대화
- 다양한 관점에서 데이터를 이해하고, 데이터 분석 방향 설정
- 데이터 구조, 중요 변수, 이상치(outlier) 파악 및 가설 검증 활용
- 가정에서 시작하는 것이 아니라, 데이터로부터 연관 관계를 파악하고 적합한 모델링 방향 설정
- 탐색적 분석 방법: 문제 -> 데이터 -> 분석 -> 모델 -> 결론
    - 전통적인 분석 방법: 문제 -> 데이터 -> 모델 -> 분석 -> 결론

## Plot

**Method 1.**
plt.plot() (plt.bar(), 등등...)으로 바로 그래프 그리기. Figure, Axes 객체를 따로 생성하지 않고도 그래프를 그릴 수 있음
```
plt.plot(x,y1,'r-*',lw=1) 
plt.plot(x,y2,'b--',lw=1)  
```

**Method 2.** 
plt.figure(), fig.subplots()로 Figure, Axes 객체를 생성하고 그래프 그리기
```
fig = plt.figure()   # 직접 Figure 객체를 생성
ax = fig.subplots()   # 직접 axes를 생성
ax.plot(x,y1,'r-*',lw=1)    # 생성된 axes에 대한 plot() 멤버 직접 호출
ax.plot(x,y2,'b--',lw=1)
```

**Method 3.** plt.subplots()로 Figure, Axes를 한 번에 생성하여 그래프 그리기
```
fig,ax = plt.subplots()  # plt.subplots() 편의 함수는 Figure 객체를 생성하고 Figure.subplots()를 호출하여 리턴
ax.plot(x,y1,'r-*',lw=1)
ax.plot(x,y2,'b--',lw=1)
```

## Subplot

**Method 1.** plt.subplot()로 Figure, Axes 객체를 따로 생성하지 않고도 그래프 그리기
```
plt.subplot(2,1,1) # 2행 1열 중 첫 번째 칸
plt.plot(x,y1,'r-*',lw=1)

plt.subplot(2,1,2) # 2행 1열 중 두 번째 칸
plt.plot(x,y2,'b--o',lw=1)
```

**Method 2.** plt.figure(), fig.subplots()로 Figure, Axes 객체를 생성하고 그래프 그리기
```
fig = plt.figure()

ax = fig.subplots(2,1)
ax[0].plot(x,y1,'r-*',lw=1)
ax[1].plot(x,y2,'b--o',lw=1)
```

**Method 3.** plt.subplots()로 Figure, Axes를 한 번에 생성하여 그래프 그리기
```
fig,ax = plt.subplots(2,1)

ax[0].plot(x,y1,'r-*',lw=1)
ax[1].plot(x,y2,'b--o',lw=1)
```

### Multiple lines
linestyle 지정 방법: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
- plot([x], y, [fmt], *, data=None, **kwargs)
> plot(x, y, 'go--', linewidth=2, markersize=12) <br>
plot(x, y, color='green', marker='o', linestyle='dashed',
...      linewidth=2, markersize=12)
- Format Strings: fmt = '[marker][line][color]' (순서 상관 없고, 세 가지 모두 지정해줘야 하는 것은 아님)
>'b'    # blue markers with default shape <br>
'or'   # red circles <br>
'-g'   # green solid line <br>
'--'   # dashed line with default color <br>
'^k:'  # black triangle_up markers connected by a dotted line <br>



````python
# data
x = np.linspace(-1, 1, 50)
y1 = 2*x + 1
y2 = 2**x + 1
````

````python
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.plot(x, y2)
plt.show()
````

````python
# y축을 각자 지정해주려면?
fig = plt.figure()
ax1 = fig.subplots()
ax1.plot(x, y1, 'r--')
ax2 = ax1.twinx()
ax2.plot(x, y2)
plt.show()
````

````python
# data
np.random.seed(1)
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
d1 = np.concatenate((spread, center, flier_high, flier_low))

spread = np.random.rand(50) * 100
center = np.ones(25) * 40
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
d2 = np.concatenate((spread, center, flier_high, flier_low))

d3 = d2[::2] # d2에서 하나 걸러 하나씩만 뽑은 것
````

````python
# data
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
````

## Method 1

````python
plt.subplot(2, 2, 1)
plt.plot(x, y)

plt.subplot(2, 2, 2)
plt.plot(x, -y)

plt.subplot(2, 2, 3)
plt.plot(x, 1+y**2)

plt.subplot(2, 2, 4)
plt.plot(x, 1-y**2)

plt.show()
````

````python
# 이렇게 하면 뭐가 달라지는 것인지 확인해보기
plt.subplot(2, 2, 1)
plt.plot(x, y)

plt.subplot(2, 2, 2)
plt.plot(x, -y)

plt.subplot(2, 1, 1)
plt.plot(x, 1+y**2)

plt.subplot(2, 1, 2)
plt.plot(x, 1-y**2)

plt.show()
````

## Method 2

````python
fig = plt.figure()

ax = fig.subplots(2,1)
ax[0].plot(x,y)
ax[1].plot(x,-y)

plt.show()
````

````python
# x축을 공유하게 하려면?
fig = plt.figure()

ax = fig.subplots(2,1, sharex=True)
ax[0].plot(x,y)
ax[1].plot(x,-y)

plt.show()
````

````python
fig, ax = plt.subplots(2, 1, sharex=True)

ax[0].plot(x,y)
ax[1].plot(x,-y)

plt.show()
````

````python
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x,y)
ax[0, 1].plot(x,-y)
ax[1, 0].plot(x, 1+y**2)
ax[1, 1].plot(x, 1-y**2)

plt.show()
````

# Plotting using pandas
functions in pandas
    - plot() : 데이터 수가 많은 경우
    - plot(kind='bar') : 데이터 수가 적은 경우
    - plot(kind='hist') : 데이터 수가 많은 경우 요약
    - plot(kind='kde') : 데이터 분포 추정
    - plot(kind='box') : 데이터의 분포 요약
    - scatter(x, y, [...]) : 두 데이터 사이의 관계
